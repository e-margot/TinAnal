# Preparing json 
(id, age, has_social, gender, is_smoking, interests, photos)

```python
f = open('/content/drive/MyDrive/TinAnal/men.json', encoding='utf-8')
ph = []
interests = []
data = json.load(f)
for i in data:
    de = []
    for j in i['options']:
      de.append(j['id'])
    for n in j['selections']:
      for m in i['photos']:
        ph.append(m['filename'])
      for inter in i['interests']:
        interests.append(inter['name'])
      with open('/content/drive/MyDrive/TinAnal/men_2.json', 'r+') as file:
        file_data = json.load(file)
        if 'de_11' in de:
          if j['id'] == 'de_11':
            if not interests:
              new = {'id': i['id'], 'age': i['age'], 'has_social': i['has_social'], 'gender': i['gender'], 'is_smoke': n['name'], 'interests': ['None'],'photos': ph}
            else:
              new = {'id': i['id'], 'age': i['age'], 'has_social': i['has_social'], 'gender': i['gender'], 'is_smoke': n['name'], 'interests': interests,'photos': ph}
        else:
          if not interests:
            new = {'id': i['id'], 'age': i['age'], 'has_social': i['has_social'], 'gender': i['gender'], 'is_smoke': 'None', 'interests': ['None'], 'photos': ph}
          else:
            new = {'id': i['id'], 'age': i['age'], 'has_social': i['has_social'], 'gender': i['gender'], 'is_smoke': 'None', 'interests': interests, 'photos': ph}
        file_data.append(new)
        file.seek(0)
        json.dump(file_data, file, indent = 4, ensure_ascii=False)
        ph.clear()
        interests.clear()


f.close()
```

Apply boolean mask

```python
def boolean_df(item_lists, unique_items):
# Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for i, item in enumerate(unique_items):
        
        # Apply boolean mask
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```
